```python
from joblib import load
from sklearn.metrics import confusion_matrix, plot_confusion_matrix
from matplotlib import pyplot as plt
from sklearn import metrics
```

```python
%run pre_proc.ipynb
```

```python
fig, ax = plt.subplots(figsize=(12,10))
LR = load('models/LR.joblib')
y_train, y_val, y_test, X_train, X_val, X_test = pre_process(tracks, features_all, feature_sets['all_non-echonest'], False)
score = round(LR.score(X_test, y_test) * 100, 2)
plot_confusion_matrix(LR, X_test, y_test, display_labels=top_genres, ax=ax)
plt.xticks(rotation=90)
ax.set_title("Confusion Matrix for Logistic Regression trained on all non-echonest features. Score={}%".format(score))
plt.show()
```

```python
y_pred = LR.predict(X_test)
acc = metrics.accuracy_score(y_test, y_pred)
recall = metrics.recall_score(y_test, y_pred, average=None)
precision = metrics.precision_score(y_test, y_pred, average=None, zero_division=0)

recall = dict(zip(top_genres, recall))
for k, v in recall.items():
    recall[k] = round(v * 100, 2)
precision = dict(zip(top_genres, precision))
for k, v in precision.items():
    precision[k] = round(v * 100, 2)
fig, ax = plt.subplots(1, 2, figsize=(12, 10))
ax[0].bar(list(recall.keys()), list(recall.values()))
ax[0].set_title("Recall Scores")
ax[1].bar(list(precision.keys()), list(precision.values()))
ax[1].set_title("Precision Scores")
plt.sca(ax[0])
plt.xticks(rotation=90)
plt.sca(ax[1])
plt.xticks(rotation=90)
plt.show()
```
